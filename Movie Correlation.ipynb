```python
# Importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# reading in the data
df = pd.read_csv(r"C:\Users\abhay sharma\Downloads\movies.csv")
```

```python
# looking at the data
df.head()
```

```python
# checking if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# looking at data types for all columns

df.dtypes
```

```python
# checking number of null values

df.isnull().sum() 
```

```python
# changing the data type of specified columns

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype('int64')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype('int64') 
```

```python
df
```

```python
# creating correct year column

df['yearcorrect'] = df['released'].str.split('[(,\s)]{1}').str[3]
```

```python
df.head()
```

```python
# sorting values

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# changing default view to display all records

pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# looking at data after sorting

df['name'].sort_values(ascending=False)
```

```python
# scatter plot - budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show
```
